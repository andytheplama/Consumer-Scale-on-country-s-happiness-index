```python
# Run this if openpyxl is not installed: pip install openpyxl
#import items
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#The Hypothesis is a product’s price will affect the quantity at which it is bought. 
# Specifically, products that are cheaper are more 
#likely to be ordered in higher quantities than products that are more expensive
# By the happiness of certain countries
```

```python
#Notes: WHR25 and figure out the top 5 happiest countries\
#Notes: then in the online_retail dataset, focus on these 5 countries

#loading the data in 
happyworld = pd.read_excel('WHR25_Data_Figure_2.1v3.xlsx')
#Inspect Column names
print(happyworld.columns)

# Sorting and selecting the top 5 happiest countries
top5 = happyworld.sort_values(by="Happiness score", ascending=False).head(5)
print(top5[["Country", "Happiness score"]])
```

```python
#Load the Data
retail = pd.read_excel("Online Retail.xlsx")

#Clean the data
retail_clean = retail[(retail["Quantity"] > 0) & (retail["UnitPrice"] > 0)]
```

```python
#Making the graphs
plt.figure(figsize=(10,6))
sns.scatterplot(data=retail_clean, x="UnitPrice", y="Quantity", alpha=0.3)
plt.xscale("log")
plt.yscale("log")
plt.title("Price vs Quantity Ordered")
plt.xlabel("Unit Price (log scale)")
plt.ylabel("Quantity Ordered (log scale)")
plt.show()


```

```python
top5 = happyworld.sort_values(by="Happiness score", ascending=False).head(5)
top5_countries = top5["Country"].tolist()

```

```python


#Filter retail data for those countries
retail_top5 = retail_clean[retail_clean["Country"].isin(top5_countries)]

#Aggregate total quantity ordered per country
country_quantity = retail_top5.groupby("Country")["Quantity"].sum().reset_index()

#Plot bar chart
plt.figure(figsize=(8,6))
plt.bar(country_quantity["Country"], country_quantity["Quantity"], color="skyblue")
plt.title("Total Quantity Ordered in Top 5 Happiest Countries")
plt.xlabel("Country")
plt.ylabel("Total Quantity Ordered")
plt.xticks(rotation=45)
plt.show()

```

```python
bottom5 = happyworld.sort_values(by="Happiness score", ascending=True).head(5)

#Filter retail data for those countries ---
bottom5_countries = bottom5["Country"].tolist()

retail_bottom5 = retail_clean[retail_clean["Country"].isin(bottom5_countries)]

#Aggregate total quantity ordered per country ---
country_quantity = retail_bottom5.groupby("Country")["Quantity"].sum().reset_index()

#Merge happiness scores with retail quantities ---
merged = pd.merge(bottom5[["Country", "Happiness score"]],
                  country_quantity,
                  on="Country",
                  how="inner")

#Plot dual-axis chart ---
fig, ax1 = plt.subplots(figsize=(10,6))

#Bar chart for total quantity ordered
ax1.bar(merged["Country"], merged["Quantity"], color="salmon", label="Total Quantity Ordered")
ax1.set_xlabel("Country")
ax1.set_ylabel("Total Quantity Ordered", color="salmon")
ax1.tick_params(axis="y", labelcolor="salmon")

#Line chart for happiness scores (secondary axis)
ax2 = ax1.twinx()
ax2.plot(merged["Country"], merged["Happiness score"],
         color="blue", marker="o", label="Happiness Score")
ax2.set_ylabel("Happiness Score", color="blue")
ax2.tick_params(axis="y", labelcolor="blue")

plt.title("Top 5 Unhappiest Countries in Retail Dataset: Happiness vs Quantity Ordered")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
score_col = "Ladder score"   # change if your column name is different

#Get the Top 5 Unhappiest Countries
unhappy5 = happyworld.sort_values(by=score_col, ascending=True).head(5)
unhappy_list = unhappy5["Country"].tolist()

print("Top 5 Unhappiest Countries:")
print(unhappy5[["Country", score_col]])

#Load Online Retail dataset
retail = pd.read_excel("Online Retail.xlsx")

# Filter for the unhappy countries
filtered = retail[retail["Country"].isin(unhappy_list)]

#Compute total quantity ordered
total_qty = filtered.groupby("Country")["Quantity"].sum().reset_index()

#Match the display order in your graph
desired_order = ["Cyprus", "France", "Greece", "Japan", "Portugal"]

total_qty = total_qty.set_index("Country").loc[desired_order].reset_index()

#Plot the bar chart
plt.figure(figsize=(10,6))
bars = plt.bar(total_qty["Country"], total_qty["Quantity"], color="darkred")

plt.title("Total Quantity Ordered in Top 5 Unhappiest Countries")
plt.xlabel("Country")
plt.ylabel("Total Quantity Ordered")

#Add labels on top
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 5,
        f"{int(height)}",
        ha="center",
        fontsize=10
    )

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import spearmanr
correlation, p_value = spearmanr(retail_clean["UnitPrice"], retail_clean["Quantity"])
print(f"Spearman correlation: {correlation:.3f}, p-value: {p_value:.3e}")
```
